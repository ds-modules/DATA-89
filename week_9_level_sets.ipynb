### Probability Surfaces Demo (Week 9)

This demo is similar to the 3D level-sets visualizer from `level_sets_and_gradients.ipynb`, but without any gradient or tangent-plane components. It provides a clean way to explore surfaces that come from familiar probability models.

The surface dropdown includes:

- **Independent Exponential Random Variables**:  
  \(f(x,y) = e^{-(x + y)}\) for \(x > 0, y > 0\), and 0 elsewhere.
- **Independent Laplace Random Variables**:  
  \(f(x,y) = e^{-|x| - |y|}\) for all \(x, y\).
- **Independent Standard Normal Random Variables**:  
  \(f(x,y) = e^{-0.5(x^2 + y^2)}\) for all \(x, y\).
- **Bivariate Student-t**:  
  \(f(x,y) = (1 + 0.5(x^2 + y^2))^{-2}\) for all \(x, y\).

Use the **Level/Plane z** slider to move a horizontal slice through the surface, and use the **Bird's-eye view** toggle to see the level set from directly above.
