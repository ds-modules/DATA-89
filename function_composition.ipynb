## Function Composition

Build composite functions step-by-step:
1. Select inner and outer functions with their parameters
2. Click **Compose Functions** to enter composition mode
3. Adjust the x slider to see the step-by-step construction:
   - Start at $(x, 0)$
   - Go up to $(x, f_{\text{inner}}(x))$
   - Move horizontally to $(f_{\text{inner}}(x), f_{\text{inner}}(x))$ on the $y=x$ line
   - Go up/down to $(f_{\text{inner}}(x), f_{\text{outer}}(f_{\text{inner}}(x)))$
   - Move horizontally to the final point $(x, f_{\text{outer}}(f_{\text{inner}}(x)))$
4. Click **Save Point** to add the final point to your collection
5. Click **Reveal Composite** to show the full composite curve
